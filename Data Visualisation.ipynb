```python
countries = pd.read_csv('countries.csv')
```

```python
countries.head()
```

```python
#Compare population growth of UK and Greece
```

```python
uk = countries[countries.country =='United Kingdom']
```

```python
greece = countries[countries.country == 'Greece']
greece
```

### population in millions

```python
# population in millions
plt.plot(uk.year, uk.population/10**6)
plt.plot(greece.year, greece.population/10**6)
plt.legend(['UK popultation', 'Greece population'])
plt.xlabel('year')
plt.ylabel('population (million)')
plt.show()
```

## relative growth in population.

```python
relative_uk = uk.population/uk.population.iloc[0] * 100
relative_gr = greece.population/greece.population.iloc[0] * 100
```

```python
plt.plot(uk.year, relative_uk)
plt.plot(greece.year, relative_gr)
plt.legend(['UK popultation growth', 'Greece population growth'])
plt.xlabel('year')
plt.xlim(1950,2008)
plt.ylabel('population rates of growth)')
plt.show()
```

```python
plt.plot(uk.year, relative_uk, label ='UK popultation growth')
plt.plot(greece.year, relative_gr, label = 'Greece population growth' )
plt.xlabel('year')
plt.xlim(1950,2008)
plt.ylabel('population rates of growth)')
plt.legend()
plt.show()
```

## Bar Chart

```python
plt.bar(uk.year, relative_uk, label ='UK popultation growth', color = 'blue')
#plt.bar(greece.year, relative_gr, label = 'Greece population growth', color = 'c' )
plt.xlabel('year')
plt.ylabel('population rates of growth)')
plt.legend()
plt.show()
```

## Histogram

```python
plt.hist(greece.population/10**6, histtype='bar', rwidth=0.8)
plt.xlabel('millions')
plt.ylabel('bins counts')
plt.title('Histogram Example, distribution of population per year')
plt.show()
```

## Scatterplot

```python
plt.scatter(uk.population/10**6, greece.population/10**6, color='k', marker='*', s = 20)
plt.xlabel('UK population in millions')
plt.ylabel('Greece population in millions')
plt.title('Scatter plot of UK and Greece Population')
```

## Stackplots (relative percentage at the whole)

```python
#what i am doing every day example

days = [1, 2 ,3 ,4,5]

sleeping = [7,6,8, 8, 9]
cooking_eating = [3,2,1,1,0]
working = [7,9,8,8,7]
leisure = [7,6,7,7,7]

plt.plot([],[], c ='b', label = 'Sleeping',  linewidth=5)
plt.plot([],[], c ='r', label = 'Cooking eating', linewidth=5)
plt.plot([],[], c ='c', label = 'working' , linewidth=5)
plt.plot([],[], c ='k', label = 'leisure', linewidth=5)


plt.stackplot(days,sleeping, cooking_eating, working, leisure, colors=['b','r','c','k'])
plt.xlabel('days')
plt.ylabel('hours')
plt.title('my daily activity')
plt.legend()
plt.show()
```

## Pie Charts

```python
slices = [5,5,10,4]
activities = ['eating', 'sleeping', 'working', 'playing']
cols = ['c','m','r','y']
plt.pie(slices, labels = activities, colors= cols, startangle= 90, shadow=True,
        explode=(0,0,0.2,0), autopct='%1.1f%%')
plt.show()
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))

ax1.plot(uk.year, relative_uk, label ='UK popultation growth')
ax1.plot(greece.year, relative_gr, label = 'Greece population growth' )
for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)
ax1.grid(True)
ax1.set_xlabel('year')  
ax1.set_ylabel('relative population growth')
ax1.xaxis.label.set_color('r')
ax1.yaxis.label.set_color('b')
#set specific values at the ticklabels
ax1.fill_between(uk.year, relative_uk, 100, alpha=0.3)
ax1.fill_between(greece.year, relative_gr, 100,  alpha=0.2, facecolor='r')
ax1.set_yticks([100,115,125,135,145])
#ax1.ylabel('population rates of growth)')
ax1.legend()
plt.show()
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))

ax1.plot(uk.year, relative_uk, label ='UK popultation growth')
ax1.plot(greece.year, relative_gr, label = 'Greece population growth' )
for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)
ax1.grid(True)
ax1.set_xlabel('year')  
ax1.set_ylabel('relative population growth')
ax1.xaxis.label.set_color('r')
ax1.yaxis.label.set_color('b')
#change the color of spines
ax1.spines['left'].set_color('c')

#make it invisible
ax1.spines['top'].set_visible(False)

#set specific values at the ticklabels
ax1.fill_between(uk.year, relative_uk, 100, alpha=0.3)

#set a line
ax1.axhline(max(relative_uk), color='k', linewidth = 2)
ax1.fill_between(greece.year, relative_gr, 100,  alpha=0.2, facecolor='r')
ax1.set_yticks([100,115,125,135,145])
#ax1.ylabel('population rates of growth)')
ax1.legend()
plt.show()
```
